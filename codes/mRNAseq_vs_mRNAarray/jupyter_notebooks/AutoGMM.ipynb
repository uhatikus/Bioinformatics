```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
csv_file_path = 'cors_20230907.csv'

# Use pandas to read the CSV file into a DataFrame
cors = pd.read_csv(csv_file_path)
```

```python
X_cor = np.array([[float(i), 1] for i in list(cors['values'])])
X_cor
```

```python
# Generate some example data
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)
```

```python
X
```

```python
n_components = 3
auto_gmm = GaussianMixture(n_components=n_components)
```

```python
# Fit the model to the data
auto_gmm.fit(X_cor)
```

```python
# Get the estimated number of components/clusters
n_components = auto_gmm.n_components_
```

```python
# Plot the data points and cluster means
plt.scatter(X_cor[:, 0], X_cor[:, 1], c=auto_gmm.predict(X_cor), cmap='viridis')
plt.scatter(auto_gmm.means_[:, 0], auto_gmm.means_[:, 1], c='red', s=100, marker='X')
plt.title(f'AutoGMM with {n_components} components')
plt.show()
```
