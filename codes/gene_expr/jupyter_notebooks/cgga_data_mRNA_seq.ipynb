```python
genes = list(set(list(pd.read_excel('gene_list.xlsx', sheet_name=0, header=None)[0]) +\
    list(pd.read_excel('gene_list.xlsx', sheet_name=1, header=None)[0])+\
    list(pd.read_excel('gene_list.xlsx', sheet_name=2, header=None)[0])))
```

```python
print(len(genes), "genes: ", genes)
```

```python
genes_in_cgga_data_mRNAseq = [g for g in genes if g in cgga_data_mRNAseq.columns]
genes_not_in_cgga_data_mRNAseq = [g for g in genes if g not in cgga_data_mRNAseq.columns]
print(f"Not in cgga_data_mRNAseq: {genes_not_in_cgga_data_mRNAseq}")
```

```python
filtered_cgga_data_mRNAseq = cgga_data_mRNAseq[genes_in_cgga_data_mRNAseq]
# filtered_cgga_data_mRNAseq = np.exp(filtered_cgga_data_mRNAseq_pre_log)
filtered_cgga_data_mRNAseq
```

```python
cgga_data_mRNAseq_pheno = pd.read_csv("CGGA.mRNAseq_325_clinical.20200506.txt", sep='\t')
# cgga_data_mRNAseq_pheno
cgga_data_mRNAseq_pheno.set_index("CGGA_ID", inplace=True)
cgga_data_mRNAseq_pheno
```

```python
cgga_data_mRNAseq_pheno.rename(columns=lambda x: x.lower(), inplace=True)
cgga_data_mRNAseq_pheno.rename(columns={'os': 'time', 'age': 'age_at_initial_pathologic_diagnosis', "censor (alive=0; dead=1)": "alive"}, inplace=True)
# cgga_data_mRNAseq_pheno["time"] = cgga_data_pheno["time"]*30
cgga_data_mRNAseq_pheno["bcr_patient_barcode"] = cgga_data_mRNAseq_pheno.index
cgga_data_mRNAseq_pheno
```

```python
cgga_data_mRNAseq_pheno["grade"].unique()
```

```python
filtered_cgga_data_mRNAseq_with_pheno_data = pd.concat([cgga_data_mRNAseq_pheno, filtered_cgga_data_mRNAseq], axis=1)
filtered_cgga_data_mRNAseq_with_pheno_data = filtered_cgga_data_mRNAseq_with_pheno_data[(filtered_cgga_data_mRNAseq_with_pheno_data["grade"] == "WHO III") | (filtered_cgga_data_mRNAseq_with_pheno_data["grade"] == "WHO II")]
filtered_cgga_data_mRNAseq_with_pheno_data
```

```python
# Calculate medians for each column
filtered_cgga_data_mRNAseq_with_medians = filtered_cgga_data_mRNAseq_with_pheno_data.copy()
medians = filtered_cgga_data_mRNAseq.median()
cols = filtered_cgga_data_mRNAseq.columns
# Create new columns for each column in the DataFrame
for col in cols:
    new_col_name = col + 'median'
    filtered_cgga_data_mRNAseq_with_medians[new_col_name] = (filtered_cgga_data_mRNAseq[col] > medians[col]).astype(int)
filtered_cgga_data_mRNAseq_with_medians
```

```python
filtered_cgga_data_mRNAseq_with_medians.to_csv('cgga_data_mRNAseq.txt', sep='\t', index=False)
```
