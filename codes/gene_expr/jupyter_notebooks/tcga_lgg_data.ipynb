```python
genes = list(set(list(pd.read_excel('gene_list.xlsx', sheet_name=0, header=None)[0]) +\
    list(pd.read_excel('gene_list.xlsx', sheet_name=1, header=None)[0])+\
    list(pd.read_excel('gene_list.xlsx', sheet_name=2, header=None)[0])))
```

```python
print(len(genes), "genes: ", genes)
```

```python
genes_in_tcga_lgg = [g for g in genes if g in tcga_lgg_data.columns]
genes_not_in_tcga_lgg = [g for g in genes if g not in tcga_lgg_data.columns]
print(f"Not in TCGA LGG: {genes_not_in_tcga_lgg}")
```

```python
filtered_tcga_lgg_data_pre_log = tcga_lgg_data[genes_in_tcga_lgg]
filtered_tcga_lgg_data = np.exp(filtered_tcga_lgg_data_pre_log)
filtered_tcga_lgg_data
```

```python
tcga_lgg_data_pheno = pd.read_csv("2023-07-31_TCGA_LGG_pheno.txt", sep='\t')
tcga_lgg_data_pheno.set_index("Sample", inplace=True)
print(tcga_lgg_data_pheno.columns)
tcga_lgg_data_pheno
```

```python
tcga_lgg_data_pheno.rename(columns=lambda x: x.lower(), inplace=True)
tcga_lgg_data_pheno.rename(columns={'survival': 'time', 'age': 'age_at_initial_pathologic_diagnosis', "status": "alive"}, inplace=True)
tcga_lgg_data_pheno["time"] = tcga_lgg_data_pheno["time"]*30
tcga_lgg_data_pheno["bcr_patient_barcode"] = tcga_lgg_data_pheno.index
tcga_lgg_data_pheno
```

```python
filtered_tcga_lgg_with_pheno_data = pd.concat([tcga_lgg_data_pheno, filtered_tcga_lgg_data], axis=1)
filtered_tcga_lgg_with_pheno_data = filtered_tcga_lgg_with_pheno_data[filtered_tcga_lgg_with_pheno_data["grade"] != "IV"]
filtered_tcga_lgg_with_pheno_data
```

```python
# Calculate medians for each column
filtered_tcga_lgg_data_with_medians = filtered_tcga_lgg_with_pheno_data.copy()
medians = filtered_tcga_lgg_data.median()
cols = filtered_tcga_lgg_data.columns
# Create new columns for each column in the DataFrame
for col in cols:
    new_col_name = col + 'median'
    filtered_tcga_lgg_data_with_medians[new_col_name] = (filtered_tcga_lgg_data[col] > medians[col]).astype(int)
filtered_tcga_lgg_data_with_medians
```

```python
filtered_tcga_lgg_data_with_medians.to_csv('tcga_lgg_data.txt', sep='\t', index=False)


```
